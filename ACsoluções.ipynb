#**Olhos e Face**

```python
imagem = cv2.imread('/content/semsorriso.jpg')
imagem_redimensionada = cv2.resize(imagem, (600, 480))
print(imagem_redimensionada.shape)
cv2_imshow(imagem_redimensionada)
```

```python
nova_largura = 600
proporcao = 375 / 500
print(proporcao)
nova_altura = int(nova_largura / proporcao)
print(nova_altura)
```

```python
imagem_cinza = cv2.cvtColor(imagem_redimensionada, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
```

```python
detector_facial = cv2.CascadeClassifier('/content/haarcascade_frontalface_default (1).xml')
detector_olhos = cv2.CascadeClassifier('/content/haarcascade_eye (1).xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
deteccoes
```

```python
imagem = imagem_redimensionada
```

```python
deteccoes_faces = detector_facial.detectMultiScale(imagem_cinza)
for (x, y, w, h) in deteccoes_faces:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
deteccoes_olhos = detector_olhos.detectMultiScale(imagem_cinza, scaleFactor = 2.7, minNeighbors=10,minSize=(40,40), maxSize=(45,45))
for (x, y, w, h) in deteccoes_olhos:
  print(w, h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 0, 255), 2)

cv2_imshow(imagem)
```

#**Full Body**

```python
fullbody_detector = cv2.CascadeClassifier('/content/haarcascade_fullbody (2).xml')
```

```python
image = cv2.imread("/content/body.jpg")
image_redimensionada = cv2.resize(image, (300, 400))
print(image_redimensionada.shape)
cv2_imshow(image_redimensionada)
```

```python
image_cinza = cv2.cvtColor(image_redimensionada, cv2.COLOR_BGR2GRAY)
image_cinza
```

```python
detections = fullbody_detector.detectMultiScale(image_cinza, scaleFactor=1.20)
for (x, y, w, h) in detections:
    cv2.rectangle(image_redimensionada, (x, y), (x + w, y + h), (0, 255, 255), 2)
    print(w, h)
```

```python
cv2_imshow(image_redimensionada)
```

#**Smile**

```python
smile_detector = cv2.CascadeClassifier('/content/haarcascade_smile (1).xml')
```

```python
image1 = cv2.imread("/content/sorriso.jfif")
imagem_redimensionada2 = cv2.resize(image1, (600, 480))
print(imagem_redimensionada2.shape)
cv2_imshow(imagem_redimensionada2)
nova_largura = 600
proporcao = 375 / 500
print(proporcao)
nova_altura = int(nova_largura / proporcao)
print(nova_altura)
```

```python
image_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
detections = smile_detector.detectMultiScale(image_gray, scaleFactor = 1.3, minNeighbors=5, minSize=(150,70), maxSize=(180,90))
for (x, y, w, h) in detections:
  print(w,h)
  cv2.rectangle(image1, (x, y), (x + w, y + h), (0,0,255), 2)
cv2_imshow(image1)
```
